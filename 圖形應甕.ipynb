```python
import matplotlib.pyplot as pt
x=[2,6,7,8,9]
y=[7,4,6,3,2]
z=[1,2,3,4,5]
pt.plot(x,y,"^",color="red",label="Marth")
pt.plot(x,z,"--",label="April")#正方形"s"
pt.legend()
pt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/180)
z=np.cos(x*np.pi/180)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.title("Sin&Cos Wave")
pt.xlabel("Degree")
pt.ylabel("Value")
pt.plot(x,y,label="sin")
pt.plot(x,z,label="cos")
pt.legend()
pt.show()
```
